```python
import os
import pandas as pd
import requests

# Define the CSV file path
csv_file = 'files/sorted_into_catagories_43-56_data.csv'


# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file)

# Define the categories and sectors
categories = ['cv', 'interesting', 'superoutburst']


sectors_with_entries = df[df[categories].eq('v').any(axis=1)]['sector'].unique()

# Iterate over each category and sector with entries
for category in categories:
    for sector in sectors_with_entries:
        sector_group = df[(df['sector'] == sector) & (df[category] == 'v')]
        if not sector_group.empty:
            sector_folder = f'{category}/Sector {sector}'
            if not os.path.exists(sector_folder):
                os.makedirs(sector_folder)

            # Iterate over each row in the sector group
            for index, row in sector_group.iterrows():
                image_link = row['image']
                image_name = os.path.basename(image_link)
                image_path = os.path.join(sector_folder, image_name)

                # Check if the image already exists in the folder
                if not os.path.exists(image_path):
                    response = requests.get(image_link)
                    if response.status_code == 200:
                        with open(image_path, 'wb') as f:
                            f.write(response.content)
                        print(f"Image downloaded for sector '{sector}', category '{category}': {image_path}")
                    else:
                        print(f"Failed to download image for sector '{sector}', category '{category}'")
                else:
                    print(f"Image already exists in folder '{sector_folder}': {image_name}")

print("Image download process completed.")
```
